```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'date_of_reservation' is datetime
df['date_of_reservation'] = pd.to_datetime(df['date_of_reservation'], errors='coerce')

# Filter only 2015 reservations
df_2015 = df[df['date_of_reservation'].dt.year == 2015]

# Extract month number
df_2015['reservation_month'] = df_2015['date_of_reservation'].dt.month

# Prepare a full 12-month range (fill 0s if no bookings in that month)
monthly_counts_2015 = df_2015['reservation_month'].value_counts().reindex(range(1, 13), fill_value=0).sort_index()

# Plotting
plt.figure(figsize=(10, 4))
sns.barplot(x=monthly_counts_2015.index, y=monthly_counts_2015.values, palette='crest')
plt.title("Number of Hotel Bookings per Month in 2015")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.xticks(ticks=range(0, 12), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
])
plt.tight_layout()
plt.show()
```

```python
# Define a function to generate the monthly bookings plot for a given year
def plot_monthly_bookings(year, color_palette):
    df_year = df[df['date_of_reservation'].dt.year == year]
    df_year['reservation_month'] = df_year['date_of_reservation'].dt.month
    
    monthly_counts = df_year['reservation_month'].value_counts().reindex(range(1, 13), fill_value=0).sort_index()
    
    plt.figure(figsize=(10, 4))
    sns.barplot(x=monthly_counts.index, y=monthly_counts.values, palette=color_palette)
    plt.title(f"Number of Hotel Bookings per Month in {year}")
    plt.xlabel("Month")
    plt.ylabel("Number of Bookings")
    plt.xticks(ticks=range(0, 12), labels=[
        'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
        'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
    ])
    plt.tight_layout()
    plt.show()

# Generate the plots
plot_monthly_bookings(2016, 'Greens')
plot_monthly_bookings(2017, 'Purples')
plot_monthly_bookings(2018, 'Oranges')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Convert to datetime if not already done
df['date_of_reservation'] = pd.to_datetime(df['date_of_reservation'], errors='coerce')

# Step 2: Extract year and month
df['reservation_year'] = df['date_of_reservation'].dt.year
df['reservation_month'] = df['date_of_reservation'].dt.month

# Step 3: Group by year and month
monthly_trends = df.groupby(['reservation_year', 'reservation_month']).size().reset_index(name='booking_count')

# Step 4: Pivot table for plotting
pivot_trends = monthly_trends.pivot(index='reservation_month', columns='reservation_year', values='booking_count').reindex(range(1,13))

# Step 5: Plot line chart
plt.figure(figsize=(12, 6))
pivot_trends.plot(marker='o', linewidth=2)
plt.title("Booking Trends Over Time (2015–2018)\nMonthly Hotel Reservation Patterns by Year\nSeasonal Trends in Hotel Booking")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.xticks(ticks=range(1, 13), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
])
plt.grid(True)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```
